```python
label_workclass = LabelEncoder() #criando um label encoder para cada uma das variavÃ©is categoricas
label_education = LabelEncoder()
label_marital = LabelEncoder()
label_occupation = LabelEncoder()
label_relationship = LabelEncoder()
```

```python
x_censu[:, 1] = label_workclass.fit_transform(x_censu[:, 1])
x_censu[:, 3] = label_education.fit_transform(x_censu[:, 3])
x_censu[:, 5] = label_marital.fit_transform(x_censu[:, 5])
x_censu[:, 6] = label_occupation.fit_transform(x_censu[:, 6])
x_censu[:, 7] = label_relationship.fit_transform(x_censu[:, 7])
#esse procedimento pode ser feito sem a necessidade do label encoder ser definido previamente
x_censu[:, 8] = LabelEncoder().fit_transform(x_censu[:, 8]) 
x_censu[:, 9] = LabelEncoder().fit_transform(x_censu[:, 9])
x_censu[:, 13] = LabelEncoder().fit_transform(x_censu[:, 13])
```

##OneHotEncoder

```python
len(np.unique(base_censu['workclass']))
```

```python
from sklearn.preprocessing import OneHotEncoder #importando o onehot encoder
from sklearn.compose import ColumnTransformer
```

```python
pd.get_dummies(x_censu[:, 1]) #usando o get_dummies para fazer um one hot encoder 
#nesse caso nÃ£o precisei fazer o label encode antes
#mudanÃ§as nÃ£o sÃ£o refletidas no DF (ou vetor)
```

```python
pd.get_dummies(x_censu[:, 1], prefix='workclass', drop_first=True)
#nesse caso todas as novas colunas comeÃ§am com 'workclass' e a primeira coluna foi apagada
```

```python
 pd.get_dummies(x_censu[:, 3], drop_first=True)#nesse caso apaga a primeira coluna
```

```python
hotencoder = ColumnTransformer(transformers=[('OneHot', OneHotEncoder(), [1, 3, 5, 6, 7, 8, 9, 13])], remainder='passthrough')
#'OneHot' Ã© sÃ³ um nome, OneHotEncoder() Ã© oque vai ser feito nas colunas, a lista de nÃºmeros se refere as colunas que serÃ£o alteraddas e
# remainder='passthrough' ignora as colunas que nÃ£o foram especificadas na lista de nÃºmeros (sem isso elas seriam apagadas)
#esse mÃ©todo Ã© melhor para fazer o one hot encoder em varias colunas 
```

```python
x_censu = hotencoder.fit_transform(x_censu).toarray()
x_censu
```
